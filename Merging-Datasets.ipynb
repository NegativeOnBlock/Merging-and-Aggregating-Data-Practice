```python
# Having memory issues and I can't output left_merged_df properly sometimes.
left_merged_df = pd.merge(left=results, right=players, left_on='away_team', right_on='away_team')

left_merged_df.head()
```

```python
# Trying to fix memory issue.
# Display the memory usage before downcasting
print("Memory usage before downcasting:")
print(left_merged_df.memory_usage(deep=True))
```

```python
# Display the DataFrame before downcasting
print("DataFrame before downcasting:")
print(left_merged_df.dtypes)
```

```python
# Identify and downcast integer columns
integer_columns = left_merged_df.select_dtypes(include='int8').columns


# Downcast integer columns
left_merged_df[integer_columns] = left_merged_df[integer_columns].apply(pd.to_numeric, downcast='integer')
```

```python
# Display the memory usage after downcasting
print("\nMemory usage after downcasting:")
print(left_merged_df.memory_usage(deep=True))
#home_score went from 90mil to 11mil
```
